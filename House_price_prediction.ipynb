```python
import pandas as pd
df = pd.read_csv('/content/Housing.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_main = LabelEncoder()
mainroad_transformed = le_main.fit_transform(df.mainroad)
```

```python
le_guest = LabelEncoder()
guestroom_transformed = le_guest.fit_transform(df.guestroom)
```

```python
le_basement = LabelEncoder()
basement_transformed = le_basement.fit_transform(df.basement)
```

```python
le = LabelEncoder()
hot_transformed = le.fit_transform(df.hotwaterheating)
```

```python
le = LabelEncoder()
air_transformed = le.fit_transform(df.airconditioning)
```

```python
le = LabelEncoder()
area_transformed = le.fit_transform(df.prefarea)
```

```python
le = LabelEncoder()
fur_transformed = le.fit_transform(df.furnishingstatus)
```

```python
y = df[['price']]
```

```python
transformed_data = {
    'mainroad_transformed': mainroad_transformed,
    'guestroom_transformed': guestroom_transformed,
    'basement_transformed': basement_transformed,
    'hot_transformed': hot_transformed,
    'air_transformed': air_transformed,
    'area_transformed': area_transformed,
    'fur_transformed': fur_transformed
}

transformed_df = pd.DataFrame(transformed_data)


concatenated_df = pd.concat([df, transformed_df], axis=1)
```

```python
concatenated_df
```

```python
concatenated_df
```

```python
columns_list = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']
```

```python
df = concatenated_df.drop(['mainroad','guestroom','basement','hotwaterheating','airconditioning'],axis=1)
```

```python
df = df.drop('prefarea',axis=1)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Price')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(df['area'], df['price'], color='green', alpha=0.7)
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Scatter Plot of Price vs. Area')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x='bedrooms', y='price', data=df)
plt.xlabel('Bedrooms')
plt.ylabel('Price')
plt.title('Box Plot of Price by Bedrooms')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr()[['price']], annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Price with Other Features')
plt.show()
```
